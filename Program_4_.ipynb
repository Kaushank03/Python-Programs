```python
def check_attendance(roll_number, present_students):
  """
  Checks if a student is present based on their roll number.

  Args:
    roll_number: The roll number of the student to check.
    present_students: A list or set of roll numbers of students who are present.

  Returns:
    A string indicating whether the student is "Present" or "Absent".
  """
  if roll_number in present_students:
    return "Present"
  else:
    return "Absent"


present_students_list = [1, 3, 5, 7]
student_roll_number = 3
status = check_attendance(student_roll_number, present_students_list)
print(f"Student {student_roll_number} is {status}")

student_roll_number = 2
status = check_attendance(student_roll_number, present_students_list)
print(f"Student {student_roll_number} is {status}")
```

```python
import numpy as np

# Sales data for 3 weeks
sales = {
    'Store A': np.array([120, 130, 115]),
    'Store B': np.array([150, 145, 160]),
    'Store C': np.array([90, 100, 95]),
    'Store D': np.array([200, 210, 205])
}

# COGS data for 3 weeks
cogs = {
    'Store A': np.array([70, 75, 72]),
    'Store B': np.array([90, 85, 88]),
    'Store C': np.array([50, 55, 53]),
    'Store D': np.array([120, 125, 122])
}

# --- Compute the weekly gross profit for all stores ---
print("\n--- Weekly Gross Profit ---")
gross_profits = {}
for store, sales_data in sales.items():
    gross_profit = sales_data - cogs[store]
    gross_profits[store] = gross_profit
    print(f"{store} Gross Profit: {gross_profit}")

# --- Which store had the highest average gross profit across the 3 weeks? ---
print("\n--- Average Gross Profit ---")
average_gross_profits = {store: np.mean(gp) for store, gp in gross_profits.items()}
for store, avg_gp in average_gross_profits.items():
    print(f"{store} Average Gross Profit: {avg_gp:.2f}")

highest_avg_gp_store = max(average_gross_profits, key=average_gross_profits.get)
print(f"Store with highest average gross profit: {highest_avg_gp_store} ({average_gross_profits[highest_avg_gp_store]:.2f})")

# --- Calculate the percentage sales growth for each store from Week-1 to Week-3 ---
print("\n--- Percentage Sales Growth (Week 1 to Week 3) ---")
for store, sales_data in sales.items():
    week1_sales = sales_data[0]
    week3_sales = sales_data[2]
    if week1_sales != 0:
        growth = ((week3_sales - week1_sales) / week1_sales) * 100
        print(f"{store} Sales Growth: {growth:.2f}%")
    else:
        print(f"{store} Sales Growth: Cannot calculate (Week 1 sales were 0)")
```

```python
# --- JAGSoM Student Data Management (Interactive for Google Colab) ---

# List of placement-relevant courses
placement_courses = [
    "Business Analytics",
    "Digital Marketing",
    "Financial Modelling",
    "Data Science",
    "HR Analytics"
]

students = []

# Pre-defined student names
student_names = [
    "Aarav", "Meera", "Riya", "Kabir", "Sanya",
    "Ishaan", "Tanya", "Vivaan", "Nisha", "Rohan"
]

print("Placement-relevant courses:")
for c in placement_courses:
    print("-", c)

# Function to check placement readiness
def is_placement_ready(courses):
    return any(c in placement_courses for c in courses)

# Collect courses for each student
for i, name in enumerate(student_names, 1):
    print(f"\nEnter courses taken by {name} (Student {i}):")

    while True:
        courses_input = input("Courses (comma-separated): ").strip()
        courses = [c.strip().title() for c in courses_input.split(",") if c.strip()]

        if is_placement_ready(courses):
            students.append({
                "Name": name,
                "Courses Taken": courses
            })
            print(f"✅ Courses recorded: {', '.join(courses)}")
            break
        else:
            print("⚠️ Must include at least one placement-relevant course! Try again.")

print("\n✅ All 10 students added successfully!\n")

# Display all students
print("--- Student Data ---")
for student in students:
    print(f"Name: {student['Name']}, Courses: {', '.join(student['Courses Taken'])}")

# Display placement-ready status
print("\n--- Placement Readiness ---")
for student in students:
    status = "placement-ready" if is_placement_ready(student["Courses Taken"]) else "NOT placement-ready"
    print(f"{student['Name']} is {status}.")
```

```python
# Parent class
class Vehicle:
    def __init__(self, brand):
        self.brand = brand

    def honk(self):
        print("Honk Honk!")

# Car class inherits from Vehicle
class Car(Vehicle):
    def __init__(self, brand, model):
        super().__init__(brand)
        self.model = model

    def honk(self):
        print(f"The {self.brand} {self.model} goes beep!")

# Bike class inherits from Vehicle
class Bike(Vehicle):
    def honk(self):
        print(f"The {self.brand} bike goes ring ring!")

# Create objects
car = Car("Toyota", "Corolla")
bike = Bike("Yamaha")

# Call methods
car.honk()   # Output: The Toyota Corolla goes beep!
bike.honk()  # Output: The Yamaha bike goes ring ring!
```

```python

from abc import ABC, abstractmethod

class Shape(ABC):
    @abstractmethod
    def area(self):
        pass

class Rectangle(Shape):
    def __init__(self, width, height):
        self.width = width
        self.height = height

    def area(self):
        return self.width * self.height

# Create rectangle object
rect = Rectangle(10, 10)
print("Rectangle area:", rect.area())
```
