```python
# Install required packages (uncomment if running first time)
# !pip install gradio pandas plotly numpy

import gradio as gr
import pandas as pd
import numpy as np
import plotly.express as px
import plotly.graph_objects as go

# --------------------------
# Generate Dummy Data
# --------------------------

# Annual Budget Data
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
income = np.random.randint(4000, 7000, size=12)
expenses = {
    'Housing': income * 0.3,
    'Food': income * 0.15,
    'Transportation': income * 0.1,
    'Savings': income * 0.2,
    'Discretionary': income * 0.25
}
budget_df = pd.DataFrame({'Month': months, 'Income': income, **expenses})
budget_df['Total_Expenses'] = budget_df[list(expenses.keys())].sum(axis=1)
budget_df['Savings_Projection'] = budget_df['Income'] - budget_df['Total_Expenses']

# Employee Performance Data
employees = ['Alice', 'Bob', 'Charlie', 'David', 'Eva']
quarters = ['Q1', 'Q2', 'Q3', 'Q4']
performance_data = []
for emp in employees:
    for q in quarters:
        performance_data.append({
            'Employee': emp,
            'Quarter': q,
            'Performance_Score': np.random.randint(60, 100),
            'Compensation': np.random.randint(4000, 7000),
            'Productivity': np.random.uniform(0.7, 1.2)
        })
employee_df = pd.DataFrame(performance_data)

# Company Performance Data
company_data = []
for q in quarters:
    company_data.append({
        'Quarter': q,
        'Revenue': np.random.randint(100000, 500000),
        'Profit': np.random.randint(100000, 3000000),
        'Market_Share': np.random.uniform(10, 30)
    })
company_df = pd.DataFrame(company_data)

# --------------------------
# Dashboard Tab Functions
# --------------------------

def annual_budget_tab():
    fig = go.Figure()
    fig.add_trace(go.Bar(x=budget_df['Month'], y=budget_df['Income'], name='Income'))
    fig.add_trace(go.Bar(x=budget_df['Month'], y=budget_df['Total_Expenses'], name='Expenses'))
    fig.add_trace(go.Line(x=budget_df['Month'], y=budget_df['Savings_Projection'], name='Savings Projection'))
    fig.update_layout(title='Annual Budget Overview', barmode='group')
    return fig

def employee_performance_tab():
    fig = px.bar(employee_df, x='Employee', y='Performance_Score', color='Performance_Score',
                 animation_frame='Quarter', range_y=[0, 100],
                 color_continuous_scale=['red','yellow','green'], title='Quarterly Employee Performance')
    return fig

def company_performance_tab():
    fig = px.line(company_df, x='Quarter', y=['Revenue','Profit'], markers=True, title='Company Performance Trend')
    fig.update_layout(yaxis_title='Amount ($)')
    return fig

def diagnostic_analytics_tab():
    budget_df['Variance'] = budget_df['Income'] - budget_df['Total_Expenses']
    fig = px.bar(budget_df, x='Month', y='Variance', title='Budget Variance Analysis (Income - Expenses)')
    return fig

def predictive_analytics_tab():
    # Simple projection using last quarter trend
    projected_revenue = company_df['Revenue'].iloc[-1] * 1.05
    projected_profit = company_df['Profit'].iloc[-1] * 1.05
    fig = go.Figure(go.Indicator(
        mode="number+delta",
        value=projected_revenue,
        delta={'reference': company_df['Revenue'].iloc[-1], 'relative': True},
        title={"text": "Projected Next Quarter Revenue"}
    ))
    return fig

def prescriptive_analytics_tab():
    # Simple recommendations
    recs = "1. Reduce discretionary expenses by 10%\n2. Increase employee incentives for top performers\n3. Optimize marketing spend to improve profit margins"
    return recs

def contact_us_info():
    return "📞 Contact Us:\nEmail: support@finance-dashboard.com\nPhone: +1-234-567-890\nFeedback Form: www.finance-dashboard.com/feedback"

# --------------------------
# Gradio Interface
# --------------------------

with gr.Blocks() as dashboard:
    gr.Markdown("## Interactive Financial Dashboard")

    # Display Contact Info at the top of the main page
    gr.Markdown(contact_us_info())

    with gr.Tab("Annual Budget"):
        gr.Plot(annual_budget_tab)

    with gr.Tab("Employee Performance"):
        gr.Plot(employee_performance_tab)

    with gr.Tab("Company Performance"):
        gr.Plot(company_performance_tab)

    with gr.Tab("Diagnostic Analytics"):
        gr.Plot(diagnostic_analytics_tab)

    with gr.Tab("Predictive Analytics"):
        gr.Plot(predictive_analytics_tab)

    with gr.Tab("Prescriptive Analytics"):
        gr.Textbox(value=prescriptive_analytics_tab(), label="Recommendations", lines=6)

dashboard.launch()
```
