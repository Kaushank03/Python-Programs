```python
# Install necessary libraries
!pip install gradio pandas numpy plotly

import gradio as gr
import pandas as pd
import numpy as np
import plotly.express as px

# -----------------------------
# Generate Dummy Data
# -----------------------------
np.random.seed(42)

# Annual Budget
annual_budget = pd.DataFrame({
    "Department": ["HR", "Sales", "IT", "Marketing", "Operations"],
    "Allocated Budget (in lakhs)": np.random.randint(50, 200, 5),
    "Spent Budget (in lakhs)": np.random.randint(30, 180, 5)
})

# Quarterly Employee Performance
employee_perf = pd.DataFrame({
    "Employee": [f"Emp {i}" for i in range(1, 11)],
    "Q1": np.random.randint(50, 100, 10),
    "Q2": np.random.randint(50, 100, 10),
    "Q3": np.random.randint(50, 100, 10),
    "Q4": np.random.randint(50, 100, 10)
})

# Quarterly Company Performance
company_perf = pd.DataFrame({
    "Quarter": ["Q1", "Q2", "Q3", "Q4"],
    "Revenue (in lakhs)": np.random.randint(200, 500, 4),
    "Profit (in lakhs)": np.random.randint(50, 150, 4)
})

# -----------------------------
# Analytics Functions
# -----------------------------

# Tab 1: Diagnostic Analytics
def diagnostic_analytics():
    fig1 = px.bar(annual_budget, x="Department", y=["Allocated Budget (in lakhs)", "Spent Budget (in lakhs)"],
                  barmode='group', title="Annual Budget vs Spent Budget")
    fig2 = px.line(employee_perf, x="Employee", y=["Q1", "Q2", "Q3", "Q4"], title="Employee Quarterly Performance")
    return fig1, fig2

# Tab 2: Predictive Analytics
def predictive_analytics():
    future_revenue = company_perf["Revenue (in lakhs)"] * np.random.uniform(1.05, 1.15, 4)
    future_profit = company_perf["Profit (in lakhs)"] * np.random.uniform(1.05, 1.2, 4)
    future_df = pd.DataFrame({
        "Quarter": company_perf["Quarter"],
        "Predicted Revenue (in lakhs)": future_revenue.round(2),
        "Predicted Profit (in lakhs)": future_profit.round(2)
    })
    fig = px.line(future_df, x="Quarter", y=["Predicted Revenue (in lakhs)", "Predicted Profit (in lakhs)"],
                  title="Predicted Company Performance")
    return fig, future_df

# Tab 3: Prescriptive Analytics
def prescriptive_analytics():
    recommendations = {
        "Recommended Actions": [
            "Invest more in Marketing Q3",
            "Conduct employee training programs",
            "Optimize inventory to reduce cost"
        ],
        "Resource Allocation Suggestions": [
            "Assign 2 more developers to IT",
            "Increase budget for Sales incentives",
            "Hire 1 more HR personnel"
        ],
        "Performance Improvement Recommendations": [
            "Set quarterly KPIs",
            "Introduce peer-review system",
            "Regular feedback meetings"
        ]
    }
    df_recommendations = pd.DataFrame(recommendations)
    return df_recommendations

# Contact Us (integrated)
def contact_us():
    contact_info = {
        "Company": ["ABC Corp."],
        "Email": ["support@abccorp.com"],
        "Phone": ["+91-9876543210"],
        "Address": ["123 Business Street, City, Country"],
        "Feedback": ["Visit www.abccorp.com/feedback to submit feedback"]
    }
    df_contact = pd.DataFrame(contact_info)
    return df_contact

# -----------------------------
# Gradio Interface
# -----------------------------
with gr.Blocks() as demo:
    gr.Markdown("## 📊 Business Analytics Dashboard")

    with gr.Tab("Diagnostic Analytics"):
        diag_fig1, diag_fig2 = diagnostic_analytics()
        gr.Plot(diag_fig1)
        gr.Plot(diag_fig2)

    with gr.Tab("Predictive Analytics"):
        pred_fig, pred_df = predictive_analytics()
        gr.Plot(pred_fig)
        gr.DataFrame(pred_df)

    with gr.Tab("Prescriptive Analytics"):
        presc_df = prescriptive_analytics()
        gr.DataFrame(presc_df)

    # Integrated Contact Us Section (below tabs)
    gr.Markdown("### 📞 Contact Us")
    contact_df = contact_us()
    gr.DataFrame(contact_df)

demo.launch()
```
