```python
# Install required libraries
!pip install gradio plotly pandas numpy faker

import gradio as gr
import pandas as pd
import numpy as np
import plotly.express as px
from faker import Faker

fake = Faker()

# -------------------------
# Generate Dummy Data
# -------------------------

# Quarterly marketing budget
def generate_budget_data():
    categories = ['Digital', 'Traditional', 'Emerging']
    budget = np.random.randint(20000, 100000, size=len(categories))
    return pd.DataFrame({'Category': categories, 'Budget ($)': budget})

# Market share data
def generate_market_share():
    competitors = ['Your Brand', 'Competitor A', 'Competitor B', 'Competitor C']
    share = np.random.randint(10, 50, size=len(competitors))
    share = (share / share.sum() * 100).round(2)
    return pd.DataFrame({'Company': competitors, 'Market Share (%)': share})

# Historical performance
def generate_historical_performance():
    months = pd.date_range(start='2025-01-01', periods=12, freq='M')
    revenue = np.random.randint(50000, 150000, size=12)
    leads = np.random.randint(500, 2000, size=12)
    return pd.DataFrame({'Month': months, 'Revenue ($)': revenue, 'Leads': leads})

# Predictive trends
def generate_predictive_data():
    months = pd.date_range(start='2025-10-01', periods=4, freq='M')
    trend_index = np.random.uniform(0.8, 1.2, size=4)  # market trend factor
    roi = np.random.randint(10, 25, size=4)  # projected ROI %
    return pd.DataFrame({'Month': months, 'Trend Index': trend_index, 'Projected ROI (%)': roi})

# Customer engagement heatmap
def generate_engagement_heatmap():
    segments = ['Gen Z', 'Millennials', 'Gen X', 'Boomers']
    channels = ['Email', 'Social Media', 'TV', 'Events']
    data = np.random.randint(50, 500, size=(len(segments), len(channels)))
    return pd.DataFrame(data, index=segments, columns=channels)

# Contact info & insights
def generate_contact_info():
    contacts = [fake.name() for _ in range(3)]
    roles = ['Marketing Director', 'Campaign Manager', 'Data Analyst']
    emails = [fake.email() for _ in range(3)]
    df = pd.DataFrame({'Name': contacts, 'Role': roles, 'Email': emails})
    return df

# -------------------------
# Dashboard Functions
# -------------------------

def diagnostic_tab():
    budget_df = generate_budget_data()
    market_df = generate_market_share()
    hist_df = generate_historical_performance()

    budget_fig = px.pie(budget_df, names='Category', values='Budget ($)', title='Marketing Budget Allocation')
    market_fig = px.bar(market_df, x='Company', y='Market Share (%)', title='Current Market Share')
    hist_fig = px.line(hist_df, x='Month', y=['Revenue ($)', 'Leads'], title='Historical Performance')

    return budget_fig, market_fig, hist_fig

def predictive_tab():
    pred_df = generate_predictive_data()
    trend_fig = px.line(pred_df, x='Month', y='Trend Index', title='Market Trend Forecast', markers=True)
    roi_fig = px.bar(pred_df, x='Month', y='Projected ROI (%)', title='Projected ROI for Marketing Initiatives')

    return trend_fig, roi_fig

def prescriptive_tab():
    strategies = [
        "Increase digital ad spend on social media platforms.",
        "Launch targeted campaigns for Gen Z and Millennials.",
        "Optimize SEO and content marketing for emerging markets.",
        "Offer limited-time promotions to increase engagement."
    ]
    recommendations = pd.DataFrame({'Recommended Actions': strategies})
    return recommendations

def contact_tab():
    contacts = generate_contact_info()
    engagement_df = generate_engagement_heatmap()
    return contacts, engagement_df

# -------------------------
# Gradio Interface
# -------------------------

with gr.Blocks() as marketing_dashboard:

    gr.Markdown("# 📊 Marketing Analytics Dashboard")

    with gr.Tabs():
        with gr.TabItem("Diagnostic Analysis"):
            budget_plot = gr.Plot()
            market_plot = gr.Plot()
            hist_plot = gr.Plot()
            refresh_diag = gr.Button("Refresh Diagnostic Data")

            def update_diag():
                budget_fig, market_fig, hist_fig = diagnostic_tab()
                return budget_fig, market_fig, hist_fig

            refresh_diag.click(update_diag, outputs=[budget_plot, market_plot, hist_plot])

        with gr.TabItem("Predictive Analysis"):
            trend_plot = gr.Plot()
            roi_plot = gr.Plot()
            refresh_pred = gr.Button("Refresh Predictive Data")

            def update_pred():
                trend_fig, roi_fig = predictive_tab()
                return trend_fig, roi_fig

            refresh_pred.click(update_pred, outputs=[trend_plot, roi_plot])

        with gr.TabItem("Prescriptive Analysis"):
            rec_table = gr.Dataframe(headers=["Recommended Actions"])
            refresh_presc = gr.Button("Refresh Recommendations")

            def update_presc():
                return prescriptive_tab()

            refresh_presc.click(update_presc, outputs=[rec_table])

        with gr.TabItem("Contact & Insights"):
            contact_table = gr.Dataframe(headers=["Name", "Role", "Email"])
            engagement_table = gr.Dataframe(headers=['Email','Social Media','TV','Events'])
            refresh_contact = gr.Button("Refresh Contact & Engagement Data")

            def update_contact():
                contacts, engagement = contact_tab()
                return contacts, engagement

            refresh_contact.click(update_contact, outputs=[contact_table, engagement_table])

marketing_dashboard.launch()
```
